```python
import pandas as pd
import seaborn as sns

sns.set_context("notebook")

df = pd.read_csv('di.csv')

def plot(question):
    data = df.loc[df["Question"] == question]
    total = data['Count'].sum()
    g = sns.catplot(kind="bar", x="Response", y="Count", data=data, height=8)
    ax = g.ax
    ax.set_title(f'{question} (n={total})')
    for p in ax.patches:
        percentage = int(p.get_height())
        x = p.get_x() + (p.get_width() / 2)
        y = p.get_height() + 1
        ax.annotate(percentage, (x, y),ha='center')
```

```python
for c in df['Question'].unique():
    plot(c)
```
